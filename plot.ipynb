***Evolution of Radial Distribution Function Over iterations***

```python
def plot2(OrbE, Densities):
    for m in range(len(OrbE)):
        x=np.linspace(0,15,200)
        fig=plt.figure(num=0,dpi=150)
        ax=fig.add_subplot()

        ax.set_xlabel('r')
        ax.set_ylabel(r'$4\pi r^{2} |\psi(r)|^{2}$')

        for n in range(len(Densities)):
         ax.plot(x,sp.lambdify((r), Densities[n][m], modules='numpy')(x),label="ITER{}".format(n+1))
        ax.set_xlim(x[0],x[-1])
        ax.set_ylim(bottom=0)
        ax.legend(loc='upper right', fontsize='small')
        plt.show()
        plt.close(fig)
```

***Evolution of Orbital's Energies over iterations***

```python
def plot3(OrbE, OrbitalEnergies):
    for m in range(len(OrbE)):
        
        fig=plt.figure(num=0,dpi=150)
        ax=fig.add_subplot()
        ax.set_xlabel('ITER')
        ax.set_ylabel(r'$E_{{{}}} \, (\mathrm{{au}})$'.format(m))

        x=[n for n in range(len(OrbitalEnergies))]
        y=[OrbitalEnergies[n][m] for n in range(len(OrbitalEnergies))]
        
        plt.scatter(x, y, c ="pink",
            linewidths = 2,
            marker ="s",
            edgecolor ="green",
            s = 50)
        plt.show()
        plt.close(fig)
```

***Effect of confinement on Orbital Energies***

```python
def plot5(OrbE, Wavefunctions):
    x = [n + 1 for n in range(len(OrbE))]
    y1 = []
    y2 = []
    for n in range(len(OrbE)):
        E_nc, E_c = pt.E_f1(n, OrbE, Wavefunctions, delta=1.89)
        y1.append(E_nc)
        y2.append(E_c)
    fig = plt.figure(num=0, dpi=150)
    ax = fig.add_subplot()

    ax.set_xticks(x)  # Set the tick positions to match the x values
    ax.set_xticklabels([f"{label}s" for label in x])  # Format the tick labels
    ax.set_xlabel('Orbital')
    ax.set_ylabel('E')
    ax.set_title('Effect of endofullerene confinement on orbital energies')
    plt.scatter(x, y1, c="pink",
                linewidths=2,
                marker="s",
                edgecolor="green",
                s=50, label='not confined')
    plt.scatter(x, y2, c="yellow",
                linewidths=2,
                marker="^",
                edgecolor="red",
                s=100, label='confined')
    ax.legend()
    plt.show()
    plt.close(fig)
```

```python
def plot4(OrbE, Wavefunctions):
    x = [n + 1 for n in range(len(OrbE))]
    y1 = []
    y2 = []
    for n in range(len(OrbE)):
        E_nc, E_c = pt.E_f2(n, OrbE, Wavefunctions)
        y1.append(E_nc)
        y2.append(E_c)
    fig = plt.figure(num=0, dpi=150)
    ax = fig.add_subplot()

    ax.set_xticks(x)  # Set the tick positions to match the x values
    ax.set_xticklabels([f"{label}s" for label in x])  # Format the tick labels
    ax.set_xlabel('Orbital')
    ax.set_ylabel('E')
    ax.set_title('Effect of Fullerene confinement on orbital energies')
    plt.scatter(x, y1, c="pink",
                linewidths=2,
                marker="s",
                edgecolor="green",
                s=50, label='not confined')
    plt.scatter(x, y2, c="yellow",
                linewidths=2,
                marker="^",
                edgecolor="red",
                s=100, label='confined')
    ax.legend()
    plt.show()
    plt.close(fig)
```
